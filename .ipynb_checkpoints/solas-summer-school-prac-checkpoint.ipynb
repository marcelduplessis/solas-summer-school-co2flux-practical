## Calculate the air-sea CO2 flux

### The calculation of the CO2 flux between the ocean and the atmosphere needs several steps:

$\newline$

2.1 Calculate ΔpCO2 (difference between atmospheric and ocean pCO2). 


\begin{equation} \Delta pCO_2 = pCO_2(atmosphere) - pCO_2(ocean) \end{equation}

```python
delta_pco2 = df['pCO2@SST (atm)'] - df['atm pCO2 (atm)']
```

$\newline$

2.2 Calculate the Schmidt number for CO2 (The Schmidt number is the ratio of the shear component for diffusivity viscosity/density to the diffusivity for mass transfer D. It physically relates the relative thickness of the hydrodynamic layer and mass-transfer boundary layer):
    
    

\begin{equation} Sn = 2073.1 - 125.62 \times SST + 3.6276 \times SST^2 - 0.043219 \times SST^3 \end{equation} 

```python
SST = df['SST (deg C)']

Sn = 2073.1 - 125.62 * SST + 3.6276 * SST**2 - 0.043219 * SST**3
```

$\newline$

2.3 Calculate the **solubility of CO2** after  Weiss(1974) in mol/l*atm with T as the absolute temperature in Kelvin (SST + 273.15): 

\begin{equation} K_0 = exp(-58.0931 + 90.569100 \times \frac{100}{T} + 22.2940 \times log_{10}\frac{T}{100} + SSS \times (0.027766 - 0.02588 \times \frac{T}{100} + 0.0050578 * \frac{T}{100}^2)) \end{equation}

```python
SSS = df['SSS (PSU)']
T = SST + 273.15

K0 = np.exp(-58.0931 + 90.569100 * (100/T) + 22.2940 * np.log(T/100) + SSS * (0.027766 - 0.02588 * (T/100) + 0.0050578 * (T/100)**2))
```

$\newline$

2.4 Calculate the transfer coefficient k(CO2) after Nightingale, 2000

\begin{equation} k(CO_2) = (0.222 \times U_{10}^2 + 0.333 \times U_{10}) \times (Sn/600)^{-0.5} \end{equation}

where $U_{10}$ is the wind speed and $Sn$ is Schmidt number for CO2.

```python
U10 = df['wind speed 10m (m/s)']

kCO2 = (0.222 * U10**2 + 0.333 * U10) * (Sn/600)**(-0.5)
```

$\newline$

2.5 Finally calculate the CO2 flux (F(CO2)) in mmol/d*m2

\begin{equation} F(CO_2) = \Delta pCO2 \times k(CO_2) \times K_0 \times 0.24 \end{equation}

```python
co2_flux = delta_pco2 * kCO2 * K0 * 0.24
```

$\newline$

Now let's write a function to calculate the CO2 flux so that we don't have go through all the steps again:

```python
def calc_CO2_flux(pCO2_air, pCO2_sea, SST, SSS, U10):
    
    # calculate the airs-sea pco2 gradient
    delta_pco2 = pCO2_sea - pCO2_air
    
    # determine the Schmidt number
    Sn = 2073.1 - 125.62 * SST + 3.6276 * SST**2 - 0.043219 * SST**3
    
    # determine the solubility of CO2
    T = SST + 273.15
    K0 = np.exp(-58.0931 + 90.569100 * (100/T) + 22.2940 * np.log(T/100) + SSS * (0.027766 - 0.02588 * (T/100) + 0.0050578 * (T/100)**2))
    
    # calculate the gas transfer coefficient for CO2
    kCO2 = (0.222 * U10**2 + 0.333 * U10) * (Sn/600)**(-0.5)
    
    # determine the flux
    co2_flux = delta_pco2 * kCO2 * K0 * 0.24
    
    return co2_flux, kCO2, K0, Sn, delta_pco2
```

$\newline$

### Plot the CO2 gradient, CO2 flux and wind speed

```python
fig,ax=plt.subplots(3, figsize=(12, 12), sharex=True)

# plot the co2 parameteres
ax[0].plot(df['Latitude (deg N)'], delta_pco2)

ax[1].plot(df['Latitude (deg N)'], co2_flux, label='CO2 flux')

# plot the wind speed
ax[2].plot(df['Latitude (deg N)'], df['wind speed 10m (m/s)'])

# makes the plots better to look at
ax[0].set_xlim(df['Latitude (deg N)'].min(), df['Latitude (deg N)'].max())

ax[2].set_xlabel('Latitude')

ax[0].set_ylabel('$\Delta$ pCO2 (sea - air)', labelpad=10)
ax[1].set_ylabel('CO$_2$ Flux', labelpad=10)
ax[2].set_ylabel('Wind speed (m/s)', labelpad=30)

ax[0].set_ylim(-150, 100)
ax[1].set_ylim(-20, 10)
ax[2].set_ylim(0, 15)

for a in ax:
    
    a.grid(c='0.75', ls='--', lw=0.5)
    a.axhline(c='k', zorder=1, lw=0.5)
    a.xaxis.set_ticks(np.arange(-50, 60, 10))
```

# Temporal variability of CO2 flux using the BATS time series

The second part of this practical is to analyse an annual time series of CO2 flux and investigate the key drivers of variability

## Load the dataset and plot the oceans variables against time.

```python
df = pd.read_csv('../bats.csv', sep=';', encoding = 'unicode_escape')

df.keys()
```

```python
time = (df['date (dd.mm.yyyy)'] + ' ' + df['Time (HH:MM:SS)'])
time = pd.to_datetime(time, format="%d.%m.%Y %H:%M:%S")
```

```python
fig = plt.figure(figsize=(9, 9))
ax = plt.axes(projection=ccrs.Stereographic(central_longitude=-40, central_latitude=0))

ax.add_feature(cartopy.feature.OCEAN, zorder=0, color='0.95')
ax.add_feature(cartopy.feature.LAND, zorder=0, edgecolor='black', color='0.65')
ax.add_feature(cartopy.feature.RIVERS, zorder=0, color='0.95')
ax.add_feature(cartopy.feature.LAKES, zorder=0, color='0.95')

img=ax.scatter(df['Longitude (deg E)'], df['Latitude (deg N)'], 
               s=100, transform=ccrs.PlateCarree())

ax.gridlines(linewidth=0.5, linestyle='--')

ax.set_extent([-90, 30, -50, 70], crs=ccrs.PlateCarree())

ax.set_title('Bermuda Atlantic Time Series')
```

### Plot all the variables as a function of the latitude

#### Questions
1. What temporal variability can you identify?
2. Can you think of reasons for the observed temporal variability?



```python
fig,ax=plt.subplots(3, figsize=(12, 12), sharex=True)
fig.subplots_adjust(hspace=0.15)

ax[0].plot(time, df['wind speed (m/s)'])
ax[1].plot(time, df['seawater pCO2 (µatm)'], label='seawater pCO2 (µatm)')
ax[1].plot(time, df['atm pCO2 (µatm)'], label='atm pCO2 (µatm)')
ax[2].plot(time, df['seawater temperature (°C)'])
ax2 = ax[2].twinx()
ax2.plot(time, df['Salinity'], c='tab:red')

# time = df.time.astype(np.datetime64)
ax[0].set_xlim(time[0], time.values[-1])

ax[0].set_ylim(0, 20)
ax[1].set_ylim(320, 520)
ax[2].set_ylim(18, 32)
ax2.set_ylim(35.6, 37)

ax[1].legend()

ax[2].set_ylabel('Temperature ($^{\circ}$C)', c='tab:blue', labelpad=30)
ax2.set_ylabel('Salinity (PSU)', c='tab:red', rotation=-90, labelpad=40)
ax[0].set_ylabel('Wind speed (m/s)', labelpad=30)
ax[1].set_ylabel('CO$_2$ (${\u03BC}$atm)', labelpad=20)

for a in ax:
    
    a.grid(lw=0.5, ls='--')
    
ax[0].xaxis.set_major_locator(months)
ax[0].xaxis.set_major_formatter(FuncFormatter(m_fmt))

ax[2].set_xlabel('2013/14', labelpad=20)
```

## Calculate the air-sea CO2 flux


```python
co2_flux, kCO2, K0, Sn, delta_pco2 = calc_CO2_flux(df['atm pCO2 (µatm)'], df['seawater pCO2 (µatm)'], df['seawater temperature (°C)'], df['Salinity'], df['wind speed (m/s)'])
```

### Plot the CO2 gradient, CO2 flux and wind speed

#### Questions

1. Describe your observation (what was expected, what is unexpected,...)
2. Describe the role of the delta pCO2 and wind speed on the magnitude and direction of the flux

```python
fig,ax=plt.subplots(3, figsize=(12, 12), sharex=True)

# plot the co2 parameteres
ax[0].plot(time, delta_pco2)

ax[1].plot(time, co2_flux, label='CO2 flux')

# plot the wind speed
ax[2].plot(time, df['wind speed (m/s)'])

# makes the plots better to look at
ax[0].set_xlim(time[0], time.values[-1])

    
ax[0].xaxis.set_major_locator(months)
ax[0].xaxis.set_major_formatter(FuncFormatter(m_fmt))

ax[0].set_ylabel('$\Delta$ pCO2 (sea - air)', labelpad=10)
ax[1].set_ylabel('CO$_2$ Flux', labelpad=10)
ax[2].set_ylabel('Wind speed (m/s)', labelpad=30)

ax[0].set_ylim(-150, 150)
ax[1].set_ylim(-35, 30)
ax[2].set_ylim(0, 20)

for a in ax:
    
    a.grid(c='0.75', ls='--', lw=0.5)
    a.axhline(c='k', zorder=1, lw=0.5)
    
ax[2].set_xlabel('2013/14', labelpad=20)
```
